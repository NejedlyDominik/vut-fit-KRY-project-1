```python
alphabet = 'abcčdeěfghijklmnopqrřsštuvwxyzž.?-/1234567890'
cipher_file = 'cipher.txt'
login = 'xnejed09'
```

```python
def rotate_right(s, n):
    return s[len(s) - n % len(s):] + s[:len(s) - n % len(s):]
```

```python
with open(cipher_file) as f: c = f.read()
```

```python
i = c.find(login)
my_cipher_per_lns = c[i - 7:i + 1256].splitlines()
my_cipher_per_lns
```

```python
rotation_shift = int(my_cipher_per_lns[1][7:].split('.', 1)[0]) - 1
rotation_shift
```

```python
encrypted_msgs = [''.join(x.split()[1:]) for x in my_cipher_per_lns[2:-1]]
encrypted_msgs
```

```python
rotated_alphabet = rotate_right(alphabet, rotation_shift)
rotated_alphabet
```

```python
removed_substitution_msgs = [''.join([rotated_alphabet[int(msg[i:i+2]) - 1] for i in range(0, len(msg), 2)]) for msg in encrypted_msgs]
removed_substitution_msgs
```

```python
def list_to_latex_tab(l, idxs=None):
    print('\\hline')
    if idxs:
        print(' & '.join(str(i + 1) for i in idxs)  + ' \\\\')
    else:
        print(' & '.join(str(i) for i in range(1, len(l[0]) + 1))  + ' \\\\')
    print('\\hline')
    for m in l:
        print(' & '.join(m) + ' \\\\')
    print('\\hline')

list_to_latex_tab(removed_substitution_msgs)
```

```python
used_symbols = []
for m in removed_substitution_msgs:
    for ch in m:
        if not ch in used_symbols:
            used_symbols.append(ch)
used_bigrams = []
for a in bigrams:
    if a[0] in used_symbols and a[1] in used_symbols:
        used_bigrams.append(a)

used_bigrams
```

```python
bigram_mtx = {}
for bigram in used_bigrams:
    if bigram[0] in bigram_mtx:
        bigram_mtx[bigram[0]][bigram[1]] = True
    else:
        bigram_mtx[bigram[0]] = {}
        bigram_mtx[bigram[0]][bigram[1]] = True

bigram_mtx
```

```python
col_follower_idxs = [[] for _ in cols]
for i, col1 in enumerate(cols):
    for j, col2 in enumerate(cols):
        if i == j:
            continue
        follows = True
        for c1, c2 in zip(col1, col2):
            if not bigram_mtx[c1].get(c2, False):
                follows = False
                break
        if follows:
            col_follower_idxs[i].append(j)

col_follower_idxs
```

```python
initial_col_idxs = [i for i, col in enumerate(cols) if '-' not in col]
initial_col_idxs.remove(48)
initial_col_idxs.insert(0, 48)
initial_col_idxs
```

```python
def find_transposition(col_idx, max_depth, col_used_flags, transposition_idxs, phrase_idx, phrase, row):
    transposition_idxs.append(col_idx)
    col_used_flags[col_idx] = True
    if phrase_idx == None and cols[col_idx][row] == phrase[0]:
        phrase_idx = 0
    elif phrase_idx != None:
        phrase_idx += 1
    if phrase_idx != None and phrase_idx > 0 and phrase_idx < len(phrase) and cols[col_idx][row] != phrase[phrase_idx]:
       pass
    elif len(transposition_idxs) == max_depth:
        yield transposition_idxs
    else:
        for next_col_idx in col_follower_idxs[col_idx]:
            if not col_used_flags[next_col_idx]:
                yield from find_transposition(next_col_idx, max_depth, col_used_flags, transposition_idxs, phrase_idx, phrase, row)
    if phrase_idx != None and phrase_idx <= 0:
        phrase_idx = None
    elif phrase_idx != None:
        phrase_idx -= 1
    transposition_idxs.pop()
    col_used_flags[col_idx] = False

def find_transpositions(initial_col_idxs, max_depth, phrase, row):
    col_used_flags = [False for _ in cols]
    transposition_idxs = []
    phrase_idx = None
    for initial_col_idx in initial_col_idxs:
        yield from find_transposition(initial_col_idx, max_depth, col_used_flags, transposition_idxs, phrase_idx, phrase, row)
```

```python
def get_msgs(col_idxs):
    return [''.join(x) for x in zip(*[cols[i] for i in col_idxs])]
```

```python
for col_idxs in find_transpositions([48],2, 'xnejed09-nove-zpravodajstvi-od-spojencu', 1):
    for i, msg in enumerate(get_msgs(col_idxs), 1):
        print(f'{i}: {msg}')
    print('---------------------------------------------------------------')
```
